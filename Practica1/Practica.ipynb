# **Objetivo**
Construir un modelo de clasificación que prediga la especie a la que pertenece una flor a partir de la longitud y anchura del pétalo y el sépalo.

Se dispone de cuatro variables predictoras (x1 = Sepal.Length, x2 = Sepal.Width, x3 = Petal.Length, x4 = Petal.Width) y una variable de clase (y = Species). El propósito es encontrar una función f(x1, x2, x3, x4) que prediga el valor de y. Es decir, se busca una función tal que y = f(x1, x2, x3, x4). 

El árbol de clasificación es una representación de dicha función.

# **Paquetes necesarios**

*   **datasets:** contiene el conjunto de datos que se usará durante el análisis. Usar *library(help="datasets")* para listar las bases de datos disponibles
*   **dplyr:** contiene funciones para la manipulación de datos tales como filtrado por filas, selección de columnas, reordenación de filas ... Usaremos la función *sample_frac* para dividir aleatoriamente el conjunto de datos

# **Obtención de un modelo de clasificación**

Obtenemos un primer árbol de clasificación que predice la especie usando la longitud y amplitud del sépalo como variables predictoras. Usamos para ello la función *rpart*. Además, visualizamos el árbol obtenido

```r
iris.tree1 <- rpart(Species ~ Sepal.Length + Sepal.Width, iris, method = "class")
```

```r
iris.tree1
```

Ahora dibujamos el árbol de clasificación usando la función *prp* del paquete *rpart.plot*

```r
prp(iris.tree1)
```

```r
prp(iris.tree1, type = 2, extra ="auto", nn = TRUE, branch = 1, varlen = 0, yesno = 2)
```

```r
predicha1 <- predict(iris.tree1, newdata = iris, type = "class")
confusionMatrix(predicha1, iris[["Species"]])
```

Obtenemos de nuevo el árbol de clasificación, pero considerando un conjunto de entrenamiento diferente del conjunto de validación. Para ello, dividimos el conjunto de datos en dos conjuntos: uno para el entrenamiento o ajuste del modelo y otro para la validación del mismo. Además, obtenemos la matriz de confusión y algunos estadísticos asociados con el clasificador.

```r
iris_entrenamiento <- sample_frac(iris, 0.7)
iris_validacion <- setdiff(iris, iris_entrenamiento)
```

```r
iris.tree2 <- rpart(Species ~ Sepal.Length + Sepal.Width, iris_entrenamiento, method = "class")
```

```r
prp(iris.tree2, type = 2, extra ="auto", nn = TRUE, branch = 1, varlen = 0, yesno = 2)
```

```r
predicha2 <- predict(iris.tree2, newdata = iris_validacion, type = "class")
confusionMatrix(predicha2, iris_validacion[["Species"]])
```

Por último, aplicamos validación cruzada para obtener y validar el árbol de clasificación. Usamos para ello la función *train()* del paquete *caret*. Debemos especificar la variable a predecir y las variables predictoras, el conjunto de datos (*data*), el método para obtener el modelo (*rpart*) y los parámetros que controlan la ejecución de la función *train()*. En este caso, indicamos que se usará validación cruzada (*method = "cv"*) con 10 pliegues (*number = 10*).

```r
train.control <- trainControl(method = "cv", number = 10)
iris.tree3 <-  train(Species ~ Sepal.Length + Sepal.Width, data = iris, method = "rpart", trControl = train.control)
prp(iris.tree3$finalModel, type = 2, extra ="auto", nn = TRUE, branch = 1, varlen = 0, yesno = 2)

predicha3 <- predict(iris.tree3$finalModel, newdata = iris_validacion, type = "class")
confusionMatrix(predicha3, iris_validacion[["Species"]])

```
