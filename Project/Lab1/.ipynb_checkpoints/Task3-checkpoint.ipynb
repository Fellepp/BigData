```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import plotly.express as px
pd.set_option("display.max.columns", None)
#pd.set_option("display.max.rows", None)
pd.set_option("display.precision", 0)
```

```python
class dfCleaner():
    def __init__(self):
        pass
    
    def clean(self):
        players = pd.read_csv('../nba_since_1950/Players.csv')
        season_stats = pd.read_csv('../nba_since_1950/Seasons_Stats.csv')
        player_info = pd.read_csv('../nba_since_1950/player_data.csv')

        players = players[~players.Player.isnull()]
        players = players.rename(columns = {'Unnamed: 0': 'id'})
        players = players[~players.born.isnull()]

        #Cleaning season stats

        season_stats = season_stats[~season_stats.Player.isnull()]
        season_stats = season_stats.iloc[:, 1:]
        season_stats = season_stats.drop(['blanl', 'blank2'], axis=1)
        season_stats['HOF'] = 'false'
        season_stats.loc[season_stats.Player.str.contains("\*"), 'HOF'] = True
        season_stats = season_stats.assign(Player=season_stats['Player'].str.replace(r'\*', ''))
        players = players.assign(Player=players['Player'].str.replace(r'\*', ''))

        season_stats['born'] = season_stats['Year'] - season_stats['Age'] - 1

        #Cleaning player data

        player_info['id'] = player_info.index
        player_info = player_info[~player_info.birth_date.isnull()]

        #Joining players and player data to ensure they have the same info

        players_born = players[['Player', 'born', 'height', 'weight']]
        #players_born[players_born.Player == "Shaquille O'Neal"]

        for i, row in player_info.iterrows():
            player_info.loc[i, 'born'] = float(row['birth_date'].split(',')[1])

        player_info_born = player_info[['name','born']]
        player_info_born.columns = ['Player', 'born']

        born = pd.concat([players_born, player_info_born])
        born = born.drop_duplicates(subset=['Player', 'born'])
        born['id'] = born.index

        # Merging id/born and seasons played

        data = season_stats.merge(born, on=['Player', 'born'])
        data = data[data.Tm != 'TOT']

        data_season = data[['id', 'Player', 'Year']].drop_duplicates()
        data_season['season'] = data_season.groupby(['id', 'Player']).cumcount() + 1
        data = data.merge(data_season, on=['id', 'Player', 'Year'])

        #data.head()

        #data = data.rename(columns = {'Unnamed: 0': 'id'})

        #data[data.Player=='Stephen Curry']
        #season_stats.head()
        data.to_csv("PlayerStat.csv")

        df = pd.read_csv("PlayerStat.csv")
        d2f = pd.read_csv("../PlayerStatsNew.csv")

        df = df.iloc[:,1:]
        d2f = d2f.iloc[:,1:]

        concatted = pd.concat([df, d2f]).drop_duplicates(subset=['Player', 'Year', 'Tm', 'born'])

        players_born = df[['Player', 'born']]
        players2_born = d2f[['Player', 'born']]
        born = pd.concat([players_born, players2_born]).drop_duplicates(subset=['Player', 'born'])
        born['id'] = born.index
        #born

        stats = concatted.merge(born, on=['Player', 'born'])
        stats = stats[stats.Tm != 'TOT']
        stats = stats.drop(['season', 'id_x'], axis=1)
        stats = stats.rename(columns={'id_y' : 'id'})
        stats_season = stats[['id', 'Player', 'Year']].drop_duplicates().sort_values('Year')
        stats_season['season'] = stats_season.groupby(['id', 'Player']).cumcount() + 1
        stats = stats.merge(stats_season, on=['id', 'Player', 'Year'])
        #stats.groupby(stats.id)['Player'].describe()
        stats = stats.sort_values(['id', 'season'])
        stats['Year'] = stats['Year'].apply(lambda x: str(int(x)))
        stats.to_csv("stats.csv")

        #size = df[['Player', 'height', 'weight']].drop_duplicates()
        #size2 = stats[['Player', 'height', 'weight']].drop_duplicates()
        #size

        df = pd.read_csv("stats.csv")
        df2 = pd.read_csv("../PlayerStat.csv")

        players = df2[['Player', 'born', 'height', 'weight']]
        playerHeights = df[['Player', 'born', 'Year']]
        players = players.merge(playerHeights, on=['Player', 'born']).drop_duplicates()
        players = players.dropna()
        players.to_csv("player_season_info.csv")
        players
```

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import plotly.express as px

class dfPlot():
    
    def __init__(self):
        pass
    
    def getLinePlot(self, data, x_label, y_label, title, start, end):
        df = pd.read_csv(data)
        df = df[df.Year <= end]
        df = df[df.Year >= start]
        if x_label == "":
            df = df.groupby(['Year'])[y_label].sum()
            plt.figure(figsize=(15,8))
            plt.plot(df)
            plt.title(title)
            plt.grid()
            plt.show()
        else:
            df = df[df.Player == x_label].sort_values('Year')
            df['Year'] = df['Year'].apply(lambda x: str(int(x)))
            df = df[['Year', y_label]]

            plt.figure(figsize=(15,8))
            plt.plot(df['Year'], df[y_label])
            plt.title(title)
            plt.grid()
            plt.show()
    
    def getBarPlot(self, data, x_label, y_label, title, rot):
        df = pd.read_csv(data)

        df = df[[x_label, y_label]].sort_values(y_label, ascending=False).head(30)

        plt.figure(figsize=(15,8))
        plt.bar(df[x_label], df[y_label])
        plt.ylabel(y_label)
        plt.xlabel(x_label)
        plt.xticks(rotation=rot)
        plt.title(title)
        plt.show()

    def getScatterPlot(self, data, x_label, y_label, title):
        df = pd.read_csv(data)
        
        df = df[[x_label, y_label, 'Player']]
        
        plt.figure(figsize=(15,8))
        plt.scatter(df[x_label], df[y_label])
        plt.show()
```
